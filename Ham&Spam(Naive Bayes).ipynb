```python
import pandas as pd
df=pd.read_csv(r"C:\Users\nt397\Downloads\spam.csv")
df.head()
```

```python
df.groupby('Category').describe()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['Category'] = le.fit_transform(df['Category'])
```

```python
df.head()
```

```python
df = df.dropna(subset=['Message', 'Category'])  # Drops rows where either column has NA
x = df['Message']
y = df['Category']
print(len(x), len(y))
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=42
)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

v = CountVectorizer()

x_train_count = v.fit_transform(x_train.values)
x_test_count = v.transform(x_test.values)

print(x_test_count.toarray()[:3])
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(x_train_count,y_train)
```

```python
emails = ["Thanks for your subscription to Ringtone UK your mobile will be charged ÃÂ£5/month Please confirm by replying YES or NO. If you reply NO you will not be charged"]
emails_count = v.transform(emails)
prediction = model.predict(emails_count)
if prediction[0] == 0:
    print("ham")
else:
    print("spam")
```

```python
x_test_count=v.transform(x_test)
model.score(x_test_count,y_test)
```
